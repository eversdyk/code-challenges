```python
def remove_exclamations(name_series):    
    return name_series.str.replace('\!', ',').str.split(',')
```

```python
names = df[['GRANTOR.LNAME','GRANTOR.FNAME','GRANTOR.MNAME','GRANTOR.TITLE','GRANTEE.LNAME','GRANTEE.FNAME','GRANTEE.MNAME','GRANTEE.TITLE']].apply(lambda x: remove_exclamations(x), axis=1)
```

```python
names['INST'] = df['INST.NUM']
```

```python
names.set_index('INST', inplace=True)
```

```python
names.head()
```

```python
data = names.to_dict('index')
```
